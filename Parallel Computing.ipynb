In multithreading, many threads of a process are executed simultaneously and process creation in multithreading is done according to economical. A common address space is shared by all the threads.

Concurrency is an approach that is used for decreasing the response time of the system by using the single processing unit.  **Write a python program to perform an addition of an array element in seqential manner and multiprocessing(module). Also compare the time between sequential and multiprocessing.**

````python
import time
import multiprocessing as mp
import numpy as np

num = range(1, 1000001)
chunk1 = num[: len(num)//2]
chunk2 = num[len(num)//2 :]

def sequential_sum(num):
    start = time.time()
    result = 0
    for i in num:
        result = result + i
    print("sequential_time: ",time.time() - start)
    return result


def parallel_sum(chunk1, chunk2):
    start = time.time()
    p1 = mp.Process(target=sum, args=(chunk1, ))
    p2 = mp.Process(target=sum, args=(chunk2, ))
  
    p1.start()
    p2.start()   

    p1.join()   
    p2.join()
    print("parallel_time: ",time.time() - start)

print(sequential_sum(num))
parallel_sum(chunk1, chunk2)
````

**Write a program to find whether the square of a number is even or odd by using the process and pool class and also draw the all posible graphs**

````python
import time
from timeit import default_timer as timer
from multiprocessing import Pool, cpu_count, Process
import os
import math

def Square(x):
  time.sleep(1)
  sqr= x ** 2
  print("Child process id:", os.getpid())
  if x % 2 == 0:
    return 'even'
  else:
    return 'odd'

def pool():
  start = timer()
  print(f'starting computations on {cpu_count()} cores')
  squareList=[4,16,33]
  print("Length of items in squareList: ",len(squareList))
  print("Parent process id:", os.getpid())
  with Pool() as pool:
    res=list(pool.map(Square,squareList))
    print(res)
  end = timer()
  print(f'elapsed time: {end - start}')

def process():
    start = timer()
    print(f'starting computations on {cpu_count()} cores')
    squareList=[4,16,33]
    print("Length of items in squareList: ",len(squareList))
    print("Parent process id:", os.getpid())
    procs = []

    # instantiating process with arguments
    for num in squareList:
        proc = Process(target=Square, args=(num,))
        procs.append(proc)
        proc.start()

    # complete the processes
    for proc in procs:
        proc.join()
    
    end = timer()
    print(f'elapsed time: {end - start}')

print("Pool Class...")
pool()

print("\n---------------")
print("Process Class.....")
process()
````

**Q1: Write differences between pool and process class**


Ans = **Pool:**

1) When you have junk of data, you can use Pool class.

2) Only the process under executions are kept in the memory.

3) I/O operation: It waits till the I/O operation is completed & does not schedule another process. This might increase the execution time.

4) Uses FIFO scheduler.

Whether a number is odd or even using process class, pool class, multithreading class and sequencial manner and also simulate the time.


**Sequential:**

````python
import multiprocessing as mp 
import os 
import time
from timeit import default_timer as timer

def res(n):
  time.sleep(0.05)
  m=n*n
  if(m%2==0):
    print("{} squared result is even".format(n))
  else:
    print("{} squared result is odd".format(n))

if __name__ == "__main__": 
  mylist = [[5, 8, 11], [1, 4, 9, 15, 22, 30], [1, 4, 9, 15, 22, 30, 5, 8, 11]]
  
  #sequentially
  print('sequentially')
  seq_time = []
  start=timer()

  for lst in mylist:
      time.sleep(0.05)
      for i in lst:
        p=i*i
        if(p%2==0):
          print("even")
        else:
          print("odd")
      end1=timer() - start
      seq_time.append(end1)
  print('sequential time : ',seq_time)
  print('\n')
````

**Process:**

````python
import multiprocessing as mp 
import os 
import time
from timeit import default_timer as timer

print("USING PROCESS ")
process_time = []
start=timer()
  #USING PROCESS CLASS
for lst in mylist:
    p2 = []
    # instantiating process with arguments
    for num in lst:
        proc = mp.Process(target=res, args=(num,))
        p2.append(proc)
        proc.start()

    # complete the processes
    for proc in p2:
        proc.join()
    end2=timer()-start
    process_time.append(end2)
print("execution time in process class", process_time)
print("\n")
````

**Pool:**

````python
import multiprocessing as mp 
import os 
import time
from timeit import default_timer as timer

print("USING POOL ")
pool_time = []
for lst in mylist:
    start=timer()
    #USING POOL CLASS
    p1=mp.Pool().map(res, lst) 
    end3= timer() - start
    pool_time.append(end3)
print("execution time in pool class", pool_time)
print("\n")
````

**Multithreading:**

````python
import multiprocessing as mp 
import os 
import time
from timeit import default_timer as timer
import threading

thread_time = []
print("USING MULTITHREADING ")
for lst in mylist:
    thread = []
    start=timer()
    # instantiating process with arguments
    for num in lst:
        proc = threading.Thread(target=res, args=(num,))
        thread.append(proc)
        proc.start()

    # complete the processes
    for proc in thread:
        proc.join()
    end4=timer()-start
    thread_time.append(end4)
print("execution time : ", thread_time)
````

Q: What is the reason to use sleep function in python. Explain with example.

The sleep() function suspends execution of the current thread for a given number of seconds. The thread will block, but the process is still alive.

In a single threaded application, this means everything is blocked while you sleep. In a multithreaded application, only the thread you explicitly 'sleep' will block and the other threads still run within the process.

````python
import time
from threading import Thread

class worker(Thread):
    def run(self):
        for x in range(0,11):
            print(x)
            time.sleep(1)

class waiter(Thread):
    def run(self):
        for x in range(100,103):
            print(x)
            time.sleep(5)

def run():
    worker().start()
    waiter().start()

run()
````

Question 3:- write a parallel processing and multi threading program to generate the nth fibonacci number also draw the time taken and speed up graph.

````python
import time

mylist = [[1, 2, 3], [1, 2, 3, 4, 5, 6], [1, 2, 3, 4, 5, 6, 7, 8, 9]]
cache = {0: 0, 1: 1}

def fibonacci_of(n):
    time.sleep(0.01)
    if n in cache:  # Base case
        return cache[n]
    # Compute and cache the Fibonacci number
    cache[n] = fibonacci_of(n - 1) + fibonacci_of(n - 2)  # Recursive case
    return cache[n]

def fibonacci_of_(n):
    if n in cache:  # Base case
        return cache[n]
    # Compute and cache the Fibonacci number
    cache[n] = fibonacci_of(n - 1) + fibonacci_of(n - 2)  # Recursive case
    return cache[n]
````

Sequential

````python
import multiprocessing as mp 
import os 
import time
from timeit import default_timer as timer

if __name__ == "__main__": 
  #sequentially
  print('Sequentially')
  seq_time = []
  start=timer()
  facts1 = []

  for lst in mylist:
      my_facts = [fibonacci_of_(x) for x in lst]
      facts1.append(my_facts)
      end1=timer() - start
      seq_time.append(end1)
      print(my_facts)
 
  print('sequential time : ',seq_time)
  print('\n')
````

Pool

````python
import multiprocessing as mp 
import os 
import time
from timeit import default_timer as timer

print("USING POOL ")
pool_time = []
for lst in mylist:
    start=timer()
    #USING POOL CLASS
    p1=mp.Pool().map(fibonacci_of, lst) 
    end3= timer() - start
    pool_time.append(end3)
print("execution time in pool class", pool_time)
print("\n")
````

Process

````python
import multiprocessing as mp 
import os 
import time
from timeit import default_timer as timer

print("USING PROCESS ")
process_time = []
start=timer()
  #USING PROCESS CLASS
for lst in mylist:
    p2 = []
    # instantiating process with arguments
    for num in lst:
        proc = mp.Process(target=fibonacci_of_, args=(num,))
        p2.append(proc)
        proc.start()

    for proc in p2:
    # complete the processes
        proc.join()
    end2=timer()-start
    process_time.append(end2)
print("execution time in process class", process_time)
print("\n")
````

Multithreading

````python
import multiprocessing as mp 
import os 
import time
from timeit import default_timer as timer
import threading

thread_time = []
print("USING MULTITHREADING ")
for lst in mylist:
    thread = []
    start=timer()
    # instantiating process with arguments
    for num in lst:
        proc = threading.Thread(target=fibonacci_of_, args=(num,))
        thread.append(proc)
        proc.start()

    # complete the processes
    for proc in thread:
        proc.join()
    end4=timer()-start
    thread_time.append(end4)
print("execution time : ", thread_time)
````

Q-4: What is open CL and CUDA? Explain

**Open CL** :- OpenCL (Open Computing Language) is a framework for writing programs that execute across heterogeneous platforms consisting of central processing units (CPUs), graphics processing units (GPUs), digital signal processors (DSPs), field-programmable gate arrays (FPGAs) and other processors or hardware accelerators. OpenCL specifies programming languages (based on C99, C++14 and C++17) for programming these devices and application programming interfaces (APIs) to control the platform and execute programs on the compute devices. OpenCL provides a standard interface for parallel computing using task- and data-based parallelism.

**Cuda** :- CUDA (or Compute Unified Device Architecture) is a parallel computing platform and application programming interface (API) that allows software to use certain types of graphics processing unit (GPU) for general purpose processing – an approach called general-purpose computing on GPUs (GPGPU). CUDA is a software layer that gives direct access to the GPU's virtual instruction set and parallel computational elements, for the execution of compute kernels.
